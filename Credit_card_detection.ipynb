```python
#under sampling
#Build  a sample dataset containing similar distribution of normal transactions and fradulent transactions
#number of fradulant transactions=492
```

```python
legit_sample=legit.sample(492)
```

```python
legit_sample.head()
```

```python
#concatenating two Dataframes
```

```python
new_dataset=pd.concat([legit_sample,fraud],axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset.shape
```

```python
new_dataset['Class'].value_counts()
```

```python
#splitting the data into features and targets
```

```python
x=new_dataset.drop(columns=['Class'])
y=new_dataset['Class']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,stratify=y,random_state=2)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
model=LogisticRegression()
model.fit(x_train,y_train)
```

```python
x_train_prediction=model.predict(x_train)
```

```python
training_data_accuracy=accuracy_score(x_train_prediction,y_train)
```

```python
print('Training data accuracy score : ',training_data_accuracy)
```

```python
#Accuracy on testing data
```

```python
x_test_prediction=model.predict(x_test)
```

```python
testing_data_accuracy=accuracy_score(x_test_prediction,y_test)
```

```python
print('Testing data accuracy score: ',testing_data_accuracy)
```
